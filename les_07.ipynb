1.Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks): zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110], ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]. Используя математические операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату (то есть, zp - признак), а за y - значения скорингового балла (то есть, ks - целевая переменная). Произвести расчет как с использованием intercept, так и без.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
# y=a+b*x
b=(np.mean(zp*ks)-np.mean(zp)*np.mean(ks))/(np.mean(zp**2)-np.mean(zp)**2)
print('b',b)
n=len(zp)
b=(n*np.sum(zp*ks)-np.sum(zp)*np.sum(ks))/(n*np.sum(zp**2)-np.sum(zp)**2)
print('b',b)
a=np.mean(ks)-b*np.mean(zp)
print('a',a)
y_hat=a+b*zp
print(y_hat)

# Построение графика
x = np.linspace(0, n, 10)
fig, ax = plt.subplots()
ax.plot(x, ks, label = 'ks')
ax.plot(x, y_hat, label = 'y_hat')
ax.set_title("Факт ks и прогноз y_hat")
ax.legend()
ax.grid()
fig.set_figheight(5)
fig.set_figwidth(8)
plt.show()

plt.scatter(zp,ks)
plt.plot(zp,a+b*zp)
```

```python
from sklearn.metrics import mean_squared_error as msee
from math import sqrt

def mse(y,y_hat):
    return ((y-y_hat)**2).sum()/n

print('rmse       : ',sqrt(mse(ks,y_hat)))
print('rmse python: ',sqrt(msee(ks,y_hat)))
```
