В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
from math import factorial
import math
from decimal import *
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))

p1 = (combinations(5,2)/combinations(8,2) * combinations(5,1)/combinations(12,1)*combinations(7,3)/combinations(12,3))# бб-бччч
p2 = (combinations(3,2)/combinations(8,2) * combinations(5,3)/combinations(12,3)*combinations(7,1)/combinations(12,1)) # чч-бббч 
p3 = (combinations(5,1)/combinations(8,1)*combinations(3,1)/combinations(8,1) * combinations(5,2)/combinations(12,2)*combinations(7,2)/combinations(12,2))# бч-ббчч
print(p1)
print(p2)
print(p3)
print('три шара белые: ',p1+p2+p3)
```
